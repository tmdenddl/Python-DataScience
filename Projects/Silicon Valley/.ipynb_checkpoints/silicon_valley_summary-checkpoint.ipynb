```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/silicon_valley_summary.csv')

df
```

```python
manager_con = df['job_category'] == 'Managers'
male_con = df['gender'] == 'Male'
not_all = df['race_ethnicity'] != 'All'

new_df = df[manager_con & male_con & not_all]
```

```python
new_df.plot(kind='bar')
```

```python
new_df.plot(kind='bar', x='race_ethnicity',  y='count')
```
