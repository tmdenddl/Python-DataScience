```python
import pandas as pd

df = pd.read_csv('data/occupations.csv')

df
```

## Sort jobs in order of average ages

```python
occupations = df.groupby('occupation')
occupations
```

```python
occupations.mean().sort_values(by='age')
```

## Sort jobs in order of number of women

```python
## Set everyone as 1
df['gender_no'] = 1
df
```

```python
## Set males as 0
male_con = df['gender'] == 'M'
df.loc[male_con, 'gender_no'] = 0
df
```

```python
occupation = df[['occupation', 'gender_no']].groupby('occupation')
occupation
```

```python
occupation.mean().sort_values(by='gender_no', ascending=False)
```

```python
occupations['gender_no'].mean().sort_values(ascending=False)
```
