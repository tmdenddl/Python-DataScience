```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://workey.codeit.kr/orangebottle/index')

soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
phoneNum = soup.select('.phoneNum')
phoneNum
```

```python
# empty array
phone_numbers = []

for number in phoneNum:
    phone_numbers.append(number.text)
    
phone_numbers
```

```python


# 결과 출력
print(phone_numbers)
```
