```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://workey.codeit.kr/music/index')

soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
orders = soup.select('.rank__order li')
orders
```

```python
orders[0].text.split()[2]
```

```python
search_ranks = []

for list in orders:
    list_text = list.text.split()[2]
    search_ranks.append(list_text)
    
search_ranks
```

```python
# 결과 출력
print(search_ranks)
```
