```python
import pandas as pd

df = pd.read_csv('data/enrolment_1.csv')
df["status"] = "allowed"

# 조건 1
boolean1 = df["course name"] == "information technology"
boolean2 = df["year"] == 1
df.loc[boolean1 & boolean2, "status"] = "not allowed"

# 조건 2
boolean3= df["course name"] == "commerce"
boolean4= df["year"] == 4
df.loc[boolean3& boolean4, "status"] = "not allowed"

# 조건 3
allowed = df["status"] == "allowed"
course_counts = df.loc[allowed, "course name"].value_counts()
closed_courses = list(course_counts[course_counts < 5].index)
for course in closed_courses:
    df.loc[df["course name"] == course, "status"] = "not allowed"

# 정답 확인
df
```

```python
import pandas as pd

df = pd.read_csv('data/enrolment_1.csv')

df
```

```python
info_tech = (df['year'] == 1) & (df['course name'] == 'information technology')
info_tech
```

```python
commerce = (df['year'] == 4) & (df['course name'] == 'commerce')
commerce
```

```python
df['status'] = 'allowed'
df
```

```python
df.loc[info_tech | commerce, 'status'] = 'not allowed'
df
```

```python
allowed = df['status'] == 'allowed'
allowed
```

```python
course_counts = df.loc[allowed, "course name"].value_counts()
course_counts
```

```python
small_courses = course_counts < 5
small_courses
```

```python
small_courses_t = small_courses[small_courses]
small_courses_t
```

```python
indexes = small_courses_t.index
indexes
```

```python
closed_courses = list(indexes)
closed_courses
```

```python
for course in closed_courses:
    df.loc[df['course name'] == course, 'status'] = 'not allowed'

df
```
