### Remove / Insert new values

```python
## Reset
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
iphone_df
```

```python
## Setting a non-existing value adds it to the data
iphone_df.loc['iPhone XR'] = ['2018-10-26', 6.1, '3GB', 'iOS 12.0.1', 'Yes']
iphone_df
```

```python
iphone_df['제조사'] = 'Apple'
iphone_df
```

```python
iphone_df.drop('iPhone XR', axis = 'index', inplace = False)
## inplace = True means if will touch the existing DataFrame
## inplace = False means if won't touch the existing DataFrame
```

```python
iphone_df
```

```python
iphone_df.drop('iPhone XR', axis='index', inplace = True)
iphone_df
```

```python
iphone_df.drop('제조사', axis='columns', inplace=True)
iphone_df
```
