```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell' : 'U.S',
    'Apple' : 'U.S',
    'Acer' : 'Taiwan',
    'HP' : 'U.S',
    'Lenovo' : 'China',
    'Alienware' : 'U.S',
    'Microsoft' : 'U.S',
    'Asus' : 'Taiwan'
}

brand_nation
```

```python
## map the value using dictionary
df['brand'].map(brand_nation)
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
df
```

### GroupBy

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.max()
```

```python
nation_groups.mean()
```

```python
### First element of each group
nation_groups.first()
```

```python
### Last element of each group
nation_groups.last()
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```
