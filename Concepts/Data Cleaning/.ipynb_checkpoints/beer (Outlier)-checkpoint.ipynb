```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
## Shows first 25%
df['abv'].quantile(0.25)
```

```python
## Shows third 25%
df['abv'].quantile(0.75)
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
## Smaller than q1 - 1.5*iqr | bigger than q3 + 1.5*iqr == outliers
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
condition
```

```python
df[condition]
```

```python
## 'Silverback Pale Ale' is a beer but abv has been inserted incorrectly
df.loc[2250, 'abv'] = 0.055
```

```python
## Smaller than q1 - 1.5*iqr | bigger than q3 + 1.5*iqr == outliers
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
outlier_index = df[condition].index
outlier_index
```

```python
## delete outliers
df.drop(outlier_index, inplace=True)
df
```

```python
## Check
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```
