```python
import pandas as pd

df = pd.read_csv('data/attendance.csv')

df
```

```python
## find where null is
df.isnull()
```

```python
## shows how many nulls there are in each column
df.isnull().sum()
```

### Three Options

1. leave it
2. drop it
3. change it

```python
# Option 2: Drop the rows
df.dropna(inplace=True)
df
```

```python
# Reset
df = pd.read_csv('data/attendance.csv')
df
```

```python
# Option 2-1: Drop the columns
df.dropna(axis='columns')
```

```python
# Reset
df = pd.read_csv('data/attendance.csv')
df
```

```python
# Option 3: Change to 0
df.fillna(0)
```

```python
# Option 3-1: Change to mean
df.fillna(df.mean())
```

```python
# Option 3-2: Change to median
df.fillna(df.median())
```
