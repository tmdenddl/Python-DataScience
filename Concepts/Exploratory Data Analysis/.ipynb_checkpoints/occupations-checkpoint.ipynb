```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/occupations.csv')
df.head()
```

### 상위 직종 3개를 골라 보세요.

```python
women = df[df['gender'] == 'F']
women
```

```python
women['occupation']
```

```python
women['occupation'].value_counts().sort_values(ascending=False)
```
